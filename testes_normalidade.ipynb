```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Se os dados são normais, usamos a estatística paramétrica. Caso contrário (qualquer outra distribuiçãou ou distribuição desconhecida), utiliza-se a estatística não-paramétrica
```

```python
from scipy import stats
from scipy.stats import skewnorm
```

```python
dados_normais = stats.norm.rvs(size=1000)
```

```python
dados_nao_normais = skewnorm.rvs(a=-10, size=1000)
```

Histograma

```python
sns.distplot(dados_normais)
```

```python
sns.distplot(dados_nao_normais)
```

Quantile-quantile plot

```python
from statsmodels.graphics.gofplots import qqplot
```

```python
qqplot(dados_normais, line='s')
```

```python
qqplot(dados_nao_normais, line='s')
```

Teste de Shapiro-Wilk

- p-value é usado para interpretar o teste estatístico
- p <= alpha: rejetita a hipótese nula, não é normal
- p > alpha: aceita a hipótese nula, é normal

```python
from scipy.stats import shapiro
```

```python
_, p = shapiro(dados_normais)
```

```python
p
```

```python
alpha = 0.05

if p > alpha:
    print('Distribuição normal')
else:
    print('Distribuição não normal')
```

```python
_, p = shapiro(dados_nao_normais)
```

```python
alpha = 0.05

if p > alpha:
    print('Distribuição normal')
else:
    print('Distribuição não normal')
```
